```python
import pandas as pd
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By
import time
```

```python
filename = pd.read_csv("corporate names ch maria.csv")
links = filename['Link']
```

```python
# driver = uc.Chrome()
```

```python
mailing_list = []
agent_list = []
officer_list = []
for i, l in enumerate(links):
    try:
        driver.get(l)
        time.sleep(2)
        try:
            mailing_address = driver.find_element(By.CSS_SELECTOR, 'div.searchResultDetail > div.detailSection:nth-child(5)').text
            if "Mailing Address" in mailing_address:
                mailing_address = mailing_address.split("Mailing Address\n")[-1].replace('\n', ' - ')
            else:
                mailing_address = None
        except:
            mailing_address = None
        try:
            reg_agent_name_and_address = driver.find_element(By.CSS_SELECTOR, 'div.searchResultDetail > div.detailSection:nth-child(6)').text
            if "Registered Agent Name & Address" in reg_agent_name_and_address:
                reg_agent_name_and_address = reg_agent_name_and_address.split("Registered Agent Name & Address\n")[-1].replace("\n", ' - ')
            else:
                reg_agent_name_and_address = None
        except:
            reg_agent_name_and_address = None
        try:
            officer_director_detail = driver.find_element(By.CSS_SELECTOR, 'div.searchResultDetail > div.detailSection:nth-child(7)').text
            if "Officer/Director Detail" in officer_director_detail:
                officer_director_detail = officer_director_detail.split("Name & Address\n\n")[-1].replace("\n", ' - ')
            else:
                officer_director_detail = None
        except:
            officer_director_detail = None

        mailing_list.append(mailing_address)
        agent_list.append(reg_agent_name_and_address)
        officer_list.append(officer_director_detail)

        print("link Number", i)    
        print(mailing_address)
        print(reg_agent_name_and_address)
        print(officer_director_detail)
    except Exception as e:
        print(e)
        filename["Mailing Address"] = mailing_list
        filename["Agent Name and Address"] = agent_list
        filename['Officer/Director Detail'] = officer_list
        
        df = pd.DataFrame(filename)
        df.to_csv('full data.csv', index=False)

filename["Mailing Address"] = mailing_list
filename["Agent Name and Address"] = agent_list
filename['Officer/Director Detail'] = officer_list

df = pd.DataFrame(filename)
df.to_csv('full data.csv', index=False)
```
